```python
def qsort(a):
    if len(a) > 1:
        pivot = a[0]
        less = a[a < pivot]
        equal = a[a == pivot]
        greater = a[a > pivot]
        return np.concatenate((qsort(less), equal, qsort(greater)))
    else:
        return a

a = np.array([5, 4, 3, 2, 1, 100], dtype=int)
qsort(a)
```
